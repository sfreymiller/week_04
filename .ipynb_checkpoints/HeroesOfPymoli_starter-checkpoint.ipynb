## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# bin the dataframe using the above method, but without dropping doubled SNs - look at all purchases
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 1000] #use existing bins
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
# bin the Age column and append to purchase_data
purchase_data['Age Category'] = pd.cut(purchase_data["Age"], age_bins, labels=age_labels)
purchase_data.head() 
age_purchase = purchase_data.groupby('Age Category') #group by Age Category

# run these calculations like with gender groupings
pch_count = age_purchase['Item ID'].count()
avg_price = round((age_purchase['Price'].mean()),2)
total_price = (age_purchase['Price'].sum())
avg_ppp = round((total_price/pch_count),2)
avg_price_per = {"Average Price Per Person": avg_ppp} #creating a DataFrame to merge
av_pp = pd.DataFrame(avg_price_per)

# merge dataframes on Age Category
inner_merge_one = pd.merge(pch_count, avg_price, on="Age Category")
inner_merge_two = pd.merge(inner_merge_one, total_price, on="Age Category")
inner_merge_three = pd.merge(inner_merge_two, av_pp, on="Age Category") 

#format headings
inner_merge_three.rename(columns = {'Item ID': 'Count', 'Price_x': 'Average Purchase Price', 'Price_y': 'Total Purchase Value', 'Average Price Per Person': 'Avg Total Purchase per Person'}, inplace = True)
inner_merge_three

#come back to add $ to values!
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group by SN
#calculate purchase count, average purchase price, total purchase value
#merge into a dataframe
#format
#sort by highest purchase count
#show the first five values using .head()

spenders = purchase_data.groupby('SN')
purchase_count = spenders['Item ID'].count() #purchase count
purchase_count.head()
average_purch_price = round((spenders['Price'].mean()),2) #average purchase price 
average_purch_price.head()
total_purch_value = spenders['Price'].sum() #total purchase value (equals avg price if only bought 1 item)
total_purch_value.head()

merge_one = pd.merge(purchase_count, average_purch_price, on='SN') #perform an inner merge
merge_two = pd.merge(merge_one, total_purch_value, on='SN')

top_spenders = merge_two.rename(columns={'Item ID': 'Purchase Count', 'Price_x': 'Average Purchase Price', 'Price_y': 'Total Purchase Value'})
top_spenders_sort = top_spenders.sort_values("Purchase Count", ascending=False)
top_spenders_sort.head() #gives top 5 values, aka a preview
```
